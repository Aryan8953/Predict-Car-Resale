```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder

model = joblib.load(r"D:\python_program\car_price_prediction\Model\random_forest_model.joblib")

df = pd.read_csv(r"D:\python_program\car_price_prediction\Dataset\car data.csv")

car_encoder = LabelEncoder()
fuel_encoder = LabelEncoder()
seller_encoder = LabelEncoder()
transmission_encoder = LabelEncoder()

car_encoder.fit(df["Car_Name"])
fuel_encoder.fit(df["Fuel_Type"])
seller_encoder.fit(df["Seller_Type"])
transmission_encoder.fit(df["Transmission"])


car_name = input("Enter Car Name: ")
year = int(input("Enter Manufacturing Year: "))
present_price = float(input("Enter Present Price (Lakhs): "))
kms_driven = int(input("Enter Kilometers Driven: "))
fuel_type = input("Enter Fuel Type: ")
seller_type = input("Enter Seller Type: ")
transmission = input("Enter Transmission: ")
owner = int(input("Enter Owner Number: "))


car_age = 2025 - year
mileage_per_year = kms_driven / (car_age + 1)

premium_brands = ["Toyota", "Honda", "Hyundai", "BMW", "Audi", "Mercedes"]

premium_brand = 1 if car_name.split()[0] in premium_brands else 0

median_kms = df["Kms_Driven"].median()
high_mileage = 1 if kms_driven > median_kms else 0


try:
    car_name = car_encoder.transform([car_name])[0]
    fuel_type = fuel_encoder.transform([fuel_type])[0]
    seller_type = seller_encoder.transform([seller_type])[0]
    transmission = transmission_encoder.transform([transmission])[0]

except ValueError:
    print("Error: Unknown category entered.")
    exit()


new_data = pd.DataFrame({
    "Car_Name":[car_name],
    "Year":[year],
    "Present_Price":[present_price],
    "Kms_Driven":[kms_driven],
    "Fuel_Type":[fuel_type],
    "Seller_Type":[seller_type],
    "Transmission":[transmission],
    "Owner":[owner],
    "Car_Age":[car_age],
    "Mileage_per_Year":[mileage_per_year],
    "Premium_Brand":[premium_brand],
    "High_Mileage":[high_mileage]
})

prediction = model.predict(new_data)

print("\nEstimated Selling Price = ₹ {:.2f} Lakhs".format(prediction[0]))
```
